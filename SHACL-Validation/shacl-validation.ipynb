```python
from pathlib import Path
from rdflib import Graph
from pyshacl import validate

DATA_DIR = Path("../")
SHAPES_FILE = "shacl-shapes.ttl"
REPORT_FILE = "validation-report.ttl"

data_graph = Graph()

for ttl_file in DATA_DIR.glob("*.ttl"):
    print(f"Lade {ttl_file}")
    data_graph.parse(ttl_file, format="turtle")

conforms, report_graph, report_text = validate(
    data_graph=data_graph,
    shacl_graph=SHAPES_FILE,
    shacl_graph_format="turtle",
    inference="rdfs",
    advanced=True,
    abort_on_first=False,
)

print(report_text)

report_graph.serialize(
    destination=REPORT_FILE,
    format="turtle"
)

print(f"Report gespeichert in: {REPORT_FILE}")

if not conforms:
    raise SystemExit(1)
```
